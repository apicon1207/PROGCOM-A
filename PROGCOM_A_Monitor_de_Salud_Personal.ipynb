#Monitor de Salud Personal

Este programa permitirá a los usuarios registrar y analizar sus signos vitales, además de calcular su frecuencia cardíaca máxima y zonas de entrenamiento.
También comparar los signos vitales con valores de referencia y da advertencias si están fuera del rango normal.


```python
def mostrar_menu():
    print("\n=== Monitor de Salud Personal ===")
    print("1. Registrar signos vitales")
    print("2. Calcular frecuencia cardíaca máxima y zonas de entrenamiento")
    print("3. Analizar signos vitales")
    print("4. Salir")

signos_vitales = []

def registrar_signos():
    frecuencia_cardiaca = int(input("Ingrese la frecuencia cardíaca (latidos por minuto): "))
    presion_arterial = input("Ingrese la presión arterial (ej. 120/80): ")
    temperatura = float(input("Ingrese la temperatura corporal (°C): "))

    signos_vitales.append((frecuencia_cardiaca, presion_arterial, temperatura))
    print("Signos vitales registrados con éxito.")

def calcular_fcm():
    edad = int(input("Ingrese su edad: "))
    fcm = 220 - edad

    zonas = {
        "Zona 1 (Reposo)": (fcm * 0.50, fcm * 0.60),
        "Zona 2 (Quema de grasa)": (fcm * 0.60, fcm * 0.70),
        "Zona 3 (Aeróbica)": (fcm * 0.70, fcm * 0.80),
        "Zona 4 (Anaeróbica)": (fcm * 0.80, fcm * 0.90),
        "Zona 5 (Máxima intensidad)": (fcm * 0.90, fcm * 1.00)
    }

    print(f"\nSu Frecuencia Cardíaca Máxima es: {fcm} latidos por minuto")
    print("\nZonas de entrenamiento:")
    for zona, (min_fc, max_fc) in zonas.items():
        print(f"{zona}: {int(min_fc)} - {int(max_fc)} latidos por minuto")

def analizar_signos():
    if not signos_vitales:
        print("No hay datos registrados.")
        return

    frecuencia_cardiaca, presion_arterial, temperatura = signos_vitales[-1]

    print("\nÚltimos signos vitales registrados:")
    print(f"Frecuencia Cardíaca: {frecuencia_cardiaca} latidos/min")
    print(f"Presión Arterial: {presion_arterial} mmHg")
    print(f"Temperatura Corporal: {temperatura}°C")

    if 60 <= frecuencia_cardiaca <= 100:
        print("✔️ Frecuencia cardíaca dentro del rango normal.")
    else:
        print("⚠️ Frecuencia cardíaca fuera del rango normal (60-100 latidos/min).")

    presion_valores = presion_arterial.split('/')
    if len(presion_valores) == 2 and presion_valores[0].isdigit() and presion_valores[1].isdigit():
        sistolica, diastolica = int(presion_valores[0]), int(presion_valores[1])
        if 90 <= sistolica <= 120 and 60 <= diastolica <= 80:
            print("✔️ Presión arterial dentro del rango normal.")
        else:
            print("⚠️ Presión arterial fuera del rango normal (90/60 - 120/80 mmHg).")
    else:
        print("⚠️ Formato de presión arterial inválido.")

    if 36.1 <= temperatura <= 37.2:
        print("✔️ Temperatura corporal dentro del rango normal.")
    else:
        print("⚠️ Temperatura fuera del rango normal (36.1°C - 37.2°C).")

def main():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            registrar_signos()
        elif opcion == "2":
            calcular_fcm()
        elif opcion == "3":
            analizar_signos()
        elif opcion == "4":
            print("Saliendo del programa. ¡Cuídate!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

if __name__ == "__main__":
    main()
```

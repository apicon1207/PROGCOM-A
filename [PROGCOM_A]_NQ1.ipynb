R/: La proteína SMN1 es un componente esencial en la biogénesis de ribonucleoproteínas nucleares pequeñas (snRNPs), fundamentales para el procesamiento del ARN mensajero (ARNm) mediante el espliceosoma. Su correcto funcionamiento es crucial para la supervivencia de neuronas motoras y propioceptivas, evitando enfermedades como la atrofia muscular espinal (AME).

Características generales:

UniProt ID: Q16637

Nombre: SMN1 (Survival Motor Neuron 1).

Gen: SMN1.

Organismo: Homo sapiens.

Longitud: 294 aminoacidos.

Peso molecular: 38 kDa

Función:
El complejo SMN ayuda con el ensamblaje de las ribonucleoproteínas nucleares pequeñas (snRNP), componente importante del espliceosoma (encargado del splicing del ARNm), fundamental para hacer un ARNm listo para traducir proteínas. Un buen funcionamiento garantiza el empalme de los genes que contienen el intrón U12 y, como consecuencia, un correcto funcionamiento de las neuronas motoras y propioceptivas. Otras dos funciones importantes son la resolución del bucle R en los híbridos ARN-ADN (si no se resuelve, el bucle R puede causar daños en el ADN) y ayudar con el metabolismo de la ribonucleoproteína nucleolar pequeña (snoRNP).

Localización:
Se encuentran en el citoplasma y en núcleo, especialmente en los GEMs (estructuras nucleares involucradas en el procesamiento del ARN y con una participación en la regulación de la proteína SMN). Otra estructura nuclear donde se pueden encontrar es en los cuerpos de Cajal (contienen junto a los GEMs la proteína SMN para madurar snRNPs). También se pueden encontrar en los gránulos citoplasmáticos, pericarion, extensiones de la membrana plasmáticas y neuronas, axón y dendritas de las neuronas motoras.

Interacciones más importantes:

	SMN-Gemin: Se compone del SMN1, GEMIN2-8 (proteinas que estabilizan y      regulan), STRAP (ayuda al reclutamiento de proteinas Sm), GEMIN3 (helicasa de ARN involucradas en el procesamiento de snRNPs, COILINA ( proteina de los cuerpo de Cajal).

	Con proteinas Sm: El SMN1 coloca correctamente las proteinas Sm para formar un anillo alrededor de los snRNAs para un correcto splicing del ARNm.

	FMRP: Para el transporte de ARN en neuronas.

	ZPR1: Localizacion nuclear de SMN.

	HNRNP: Participa en el procesamiento de ARN.

	PLGR1 y PRMT5: Modifican SMN.

Vias metabolicas:

	Biogénesis de snRNPs:
1.	Se ensambla el complejo SMN-Gemin.
2.	Se reclutan proteínas Sm.
3.	Se forma el anillo alrededor del snRNA.
4.	Se madura para luego la exportación nuclear de los snRNPs.
5.	Se incorpora al espliceosoma.

	Transporte axonal de ARN en neuronas motoras: SMN1 se asocia con B-actina y proteínas de transporte motor para llevar ARNm a las sinapsis neuronales

	Señalización BMP/TGF-B: SMN1 afecta esta vía para prevenir apoptosis. Una sobre activación de BMP puede empeorar el AME.

	Metilación de SMN1 en residuos de arginina.

	PRMT5 es importante para la función de SMN1.

	Fosforilación de SMN1 es regulada por CDKs y AKT, influye en la localización.

Mutaciones en SMN1 asociadas a la Atrofia Muscular Espinal (AME):
Diversas mutaciones en la proteína SMN1 afectan su funcionalidad y contribuyen al desarrollo de la Atrofia Muscular Espinal (AME), una enfermedad neurodegenerativa caracterizada por la pérdida progresiva de neuronas motoras.
Principales mutaciones de la proteína:
o	VAR_034807 (I116F)
La mutación en la posición 116 cambia una isoleucina (I) por fenilalanina (F), lo que provoca la abolición de la interacción de SMN con la proteína ELAVL4.  Esta interacción es clave para la regulación y estabilidad del ARN en neuronas motoras, por lo que su pérdida puede contribuir a la degeneración neuronal.
o	VAR_034808 (Q136E)
La mutación en la posición 136 cambia una glutamina (Q) por ácido glutámico (E). Este cambio interfiere con la unión de SMN a sustratos que contienen arginina dimetilada y también afecta su interacción con ELAVL4. Esto altera la formación de complejos ribonucleoproteicos esenciales, como los snRNPs, que participan en el proceso de splicing del ARN, afectando la maduración del ARN mensajero en las neuronas.
o	VAR_005617 (Y272C)
La mutación en la posición 272 cambia una tirosina (Y) por cisteína (C), lo que hace que ocurra la abolición de la unión de SMN a RPP20/POP7 y una reducción significativa en su unión con SNRPB y GEMIN8, además de afectar su capacidad de homooligomerización. Estas alteraciones comprometen la formación del complejo SMN, afectando su función en el ensamblaje de ribonucleoproteínas esenciales para la actividad neuronal.
o	VAR_005620 (G279V)
La mutación en la posición 279 cambia una glicina (G) por valina (V). Esto disminuye la capacidad de SMN para unirse a RPP20/POP7 y afecta su homooligomerización, lo que impide la formación eficiente del complejo SMN. Además, altera su distribución en los axones neuronales, afectando el transporte de ARN y proteínas necesarias para la función y mantenimiento de las neuronas motoras.

```python
import random
import time
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output, Image

opciones = {
    "Piedra": "https://png.pngtree.com/png-clipart/20220509/original/pngtree-fist-vector-illustration-png-image_7670867.png",
    "Papel": "https://cdn-icons-png.flaticon.com/512/9125/9125112.png",
    "Tijera": "https://w7.pngwing.com/pngs/534/90/png-transparent-finger-thumb-scissor-hand-logo-art.png",
    "Spock": "https://e7.pngegg.com/pngimages/474/918/png-clipart-vulcan-salute-spock-emoji-star-trek-emoji-hand-salute.png",
    "Lagarto": "https://images.vexels.com/content/208908/preview/lizard-silhouette-top-view-b5d823.png",
    "Fuego": "https://png.pngtree.com/png-clipart/20210704/original/pngtree-fuego-png-png-image_6502195.jpg",
    "Agua": "https://png.pngtree.com/png-clipart/20220117/original/pngtree-water-drop-cartoon-hand-drawn-elements-png-image_7131235.png",
    "Aire": "https://w7.pngwing.com/pngs/536/640/png-transparent-air-computer-icons-classical-element-symbol-air-miscellaneous-text-heart.png",
    "Rayo": "https://i.pinimg.com/736x/dd/c3/9e/ddc39eacec124b7e82b61a6fba8cfa9b.jpg"
}

reglas = {
    "Piedra": ["Tijera", "Lagarto", "Fuego"],
    "Papel": ["Piedra", "Spock", "Agua"],
    "Tijera": ["Papel", "Lagarto", "Aire"],
    "Spock": ["Tijera", "Piedra", "Rayo"],
    "Lagarto": ["Papel", "Spock", "Aire"],
    "Fuego": ["Papel", "Lagarto", "Rayo"],
    "Agua": ["Fuego", "Piedra", "Rayo"],
    "Aire": ["Agua", "Fuego", "Piedra"],
    "Rayo": ["Spock", "Aire", "Tijera"]
}

output = widgets.Output()

def jugar(eleccion_usuario):
    clear_output()
    display(HTML('<h2 style="color: white; background-color: darkblue; padding: 10px;">Juego Piedra, Papel, Tijera (versión extendida)</h2>'))
    display(caja_botones)
    display(output)

    with output:
        clear_output(wait=True)
        print("Eligiendo...")
        time.sleep(1)
        eleccion_computadora = random.choice(list(opciones.keys()))

        print(f"Tu elección: {eleccion_usuario}")
        display(Image(url=opciones[eleccion_usuario], width=100, height=100))

        print(f"Elección de la computadora: {eleccion_computadora}")
        display(Image(url=opciones[eleccion_computadora], width=100, height=100))

        if eleccion_usuario == eleccion_computadora:
            print("¡Empate!")
        elif eleccion_computadora in reglas[eleccion_usuario]:
            print("¡Ganaste!")
        else:
            print("Perdiste :(")

botones = [widgets.Button(description=opcion) for opcion in opciones.keys()]
for boton in botones:
    boton.on_click(lambda b, opcion=boton.description: jugar(opcion))

caja_botones = widgets.HBox(botones)

display(HTML('<h2 style="color: white; background-color: darkblue; padding: 10px;">Juego Piedra, Papel, Tijera (versión extendida)</h2>'))
display(caja_botones)
display(output)
```
